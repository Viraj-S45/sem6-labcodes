```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
image_url = 'https://b2633864.smushcdn.com/2633864/wp-content/uploads/2020/12/opencv_load_image_header.jpg?size=630x356&lossy=2&strip=1&webp=1'

# Download the image
response = requests.get(image_url)
image_np = np.frombuffer(response.content, np.uint8)

# Read image from numpy array
image = cv2.imdecode(image_np, cv2.IMREAD_COLOR)

# Show image
plt.imshow(image)
plt.axis('off')
plt.title("Original Image")
plt.show()
```

```python
# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Edge detection using Canny
edges = cv2.Canny(gray, 100, 200)

# Display results
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.show()
```

```python
# Apply Gaussian Blur
blurred = cv2.GaussianBlur(image, (15, 15), 0)

# Display
plt.figure(figsize=(6,4))
plt.imshow(blurred)
plt.axis('off')
plt.title("Blurred Image")
```

```python
# Edge detection
edges = cv2.Canny(gray, 100, 200)

# Display
plt.figure(figsize=(6,4))
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.title("Canny Edge Detection")
```

## xgboost

#### Defination:

XGBoost (Extreme Gradient Boosting) is an optimized, scalable, and high-performance machine learning library designed for **supervised learning** tasks such as classification, regression, and ranking.
It is based on the **gradient boosting framework** and builds an ensemble of decision trees sequentially, where each new tree corrects the errors made by previous trees.

#### Advantages of OpenCV

- Handles Missing Values Automatically
- Regularization Support
- Scalable and Efficient
- Built-in Cross Validation and Early Stopping


#### Disadvantages of OpenCV

- Complexity
- Memory Intensive
- Overfitting Risk
- Long Training Time for Large Models

```python
import xgboost as xgb
import numpy as np

# Data: [hours studied, attendance %, previous score]
X = np.array([
    [2, 60, 45],
    [4, 70, 55],
    [6, 80, 65],
    [8, 85, 75],
    [10, 90, 85],
    [1, 50, 40],
    [3, 65, 50],
    [7, 88, 78],
    [9, 92, 88],
    [5, 75, 60]
])

# 0 = Fail, 1 = Pass
y = np.array([0, 0, 1, 1, 1, 0, 0, 1, 1, 1])

# Create DMatrix
dtrain = xgb.DMatrix(X, label=y)

# Parameters
params = {
    'max_depth': 3,
    'eta': 0.1,
    'objective': 'binary:logistic',
    'eval_metric': 'auc'
}

# Train model
model = xgb.train(params, dtrain, num_boost_round=30)

# Predict probability of passing
predictions = model.predict(dtrain)
print([f'{val:.2f}' for val in predictions * 100])
```

```python
import xgboost as xgb
import numpy as np
from sklearn.model_selection import train_test_split

# Data: [size, bedrooms, distance from city]
X = np.array([
    [800, 1, 15],
    [1000, 2, 12],
    [1200, 2, 10],
    [1500, 3, 8],
    [1800, 3, 6],
    [2000, 4, 5],
    [2200, 4, 4],
    [2500, 5, 3]
])

# House prices (thousands)
y = np.array([50, 70, 90, 120, 150, 180, 200, 230])

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Create regressor
model = xgb.XGBRegressor(
    n_estimators=100,
    max_depth=4,
    learning_rate=0.1
)

# Train model
model.fit(X_train, y_train)

# Predict prices
predictions = model.predict(X_test)
print([f'{val:.2f}' for val in predictions])
```

## FastAI

#### Defination:

fastai is an open-source deep learning library built on top of PyTorch that simplifies the process of building, training, and deploying deep learning models. It provides high-level APIs for common deep learning tasks such as computer vision, natural language processing, tabular data, and recommendation systems, while still allowing full access to low-level PyTorch components for customization. fastai is designed to be both easy for beginners and flexible for advanced researchers, enabling rapid experimentation and state-of-the-art performance with minimal code.

#### Advantages of FastAI

- Ease of Use and Rapid Development
- High-Level Abstractions with Flexibility
- Powerful Callback System
- Colab and GPU Friendly


#### Disadvantages of FastAI

- Abstraction Can Hide Details
- Learning Curve for Advanced Customization
- Dependency on PyTorch Updates
- Smaller Ecosystem Compared to TensorFlow

```python
from fastai.vision.all import *

# Load sample dataset
path = untar_data(URLs.PETS)

# Create DataLoaders
dls = ImageDataLoaders.from_path_re(
    path, # Root path for the data
    get_image_files(path/"images"), # List of image files
    pat=r'(.+)_\d+.jpg$', # Corrected regex pattern to extract labels from filenames
    item_tfms=Resize(224) # Apply a resize transformation to each item
)

# Create a learner with a pretrained model
learn = vision_learner(dls, resnet18, metrics=accuracy)

# Train the model
learn.fine_tune(3)
```

```python
from fastai.text.all import *

# Load IMDB movie reviews dataset
path = untar_data(URLs.IMDB)

# Create DataLoaders
dls = TextDataLoaders.from_folder(
    path,
    train='train',
    valid='test',
    is_lm=False
)

# Create a text learner
learn = text_classifier_learner(dls, AWD_LSTM, metrics=accuracy)

# Train the model
learn.fine_tune(2)
```

```python
from fastai.tabular.all import *

# Load sample tabular dataset
path = untar_data(URLs.ADULT_SAMPLE)
df = pd.read_csv(path/'adult.csv')

# Define preprocessing steps
procs = [Categorify, FillMissing, Normalize]
dep_var = 'salary'
cat_names = ['workclass', 'education', 'marital-status', 'occupation']
cont_names = ['age', 'education-num', 'hours-per-week']

# Create DataLoaders
dls = TabularDataLoaders.from_df(
    df,
    path=path,
    procs=procs,
    cat_names=cat_names,
    cont_names=cont_names,
    y_names=dep_var
)

# Create learner
learn = tabular_learner(dls, metrics=accuracy)

# Train the model
learn.fit_one_cycle(3)
```

## Keras

### Definition

**Keras** is a high-level Python library used for **deep learning**.
It provides a simple and user-friendly interface for building, training, and evaluating **neural networks**, and typically runs on top of **TensorFlow**.


### Advantages of Keras

1. Simple and easy-to-use API
2. Rapid model development
3. Supports CNNs, RNNs, and ANNs
4. Built on TensorFlow backend
5. Good documentation and community support

### Disadvantages of Keras

1. Less control over low-level operations
2. Slower than pure TensorFlow for custom models
3. Limited flexibility for research-level customization
4. Depends on TensorFlow backend

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(8, activation='relu', input_shape=(4,)))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```
