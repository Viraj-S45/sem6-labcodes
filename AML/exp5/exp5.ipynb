```python
# Import Libraries
import numpy as np
```

```python
# Define HMM Forward Algorithm
def forward_algorithm(states, observations, start_prob, trans_prob, emit_prob, obs_sequence):
    n_states = len(states)
    T = len(obs_sequence)

    alpha = np.zeros((T, n_states))

    # Initialization
    for i, state in enumerate(states):
        alpha[0, i] = start_prob[state] * emit_prob[state][obs_sequence[0]]

    # Recursion
    for t in range(1, T):
        for j, curr_state in enumerate(states):
            sum_prob = 0
            for i, prev_state in enumerate(states):
                sum_prob += alpha[t-1, i] * trans_prob[prev_state][curr_state]
            alpha[t, j] = sum_prob * emit_prob[curr_state][obs_sequence[t]]

    # Termination
    final_prob = np.sum(alpha[T-1])

    return final_prob, alpha
```

```python

states = ['Rainy', 'Sunny']
observations = ['Walk', 'Shop', 'Clean']

start_prob = {
    'Rainy': 0.6,
    'Sunny': 0.4
}

trans_prob = {
    'Rainy': {'Rainy': 0.7, 'Sunny': 0.3},
    'Sunny': {'Rainy': 0.4, 'Sunny': 0.6}
}

emit_prob = {
    'Rainy': {'Walk': 0.1, 'Shop': 0.4, 'Clean': 0.5},
    'Sunny': {'Walk': 0.6, 'Shop': 0.3, 'Clean': 0.1}
}
```

```python
# Sequence
obs_sequence = ['Walk', 'Shop', 'Clean']
```

```python
# Forward Algorithm
prob, alpha_matrix = forward_algorithm(
    states, observations, start_prob, trans_prob, emit_prob, obs_sequence
)

print("Alpha Matrix:\n", alpha_matrix)
print("\nFinal Probability:", prob)
```
