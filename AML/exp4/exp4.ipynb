```python
# Libraries
import numpy as np
```

```python
# Data
data = np.array([
    [3, 7],
    [4, 6],
    [7, 3],
    [9, 1]
])
```

```python
# Parameters
theta_A = 0.6
theta_B = 0.5
iteration = 0
```

```python
while True:
    iteration += 1

    old_theta_A = round(theta_A, 4)
    old_theta_B = round(theta_B, 4)

    weights_A = []
    weights_B = []

    # Expectation
    for H, T in data:

        likelihood_A = (theta_A ** H) * ((1 - theta_A) ** T)
        likelihood_B = (theta_B ** H) * ((1 - theta_B) ** T)

        prob_A = likelihood_A / (likelihood_A + likelihood_B)
        prob_B = 1 - prob_A

        weights_A.append(prob_A)
        weights_B.append(prob_B)

    weights_A = np.array(weights_A)
    weights_B = np.array(weights_B)

    # Maximization
    heads_A = np.sum(weights_A * data[:, 0])
    tails_A = np.sum(weights_A * data[:, 1])

    heads_B = np.sum(weights_B * data[:, 0])
    tails_B = np.sum(weights_B * data[:, 1])

    theta_A = heads_A / (heads_A + tails_A)
    theta_B = heads_B / (heads_B + tails_B)

    print(f"\nIteration {iteration}")
    print("Theta_A =", round(theta_A, 4))
    print("Theta_B =", round(theta_B, 4))

    if (round(theta_A, 4) == old_theta_A and
        round(theta_B, 4) == old_theta_B):
        print("\nConverged")
        break
```
